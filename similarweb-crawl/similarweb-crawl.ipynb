## Find header bidding sites

```python
df = pd.read_sql('similarweb', con=get_engine())
```

```python
hb = {}
hb_ = pd.read_sql('header-bidding-sites', con=get_engine())
for row in hb_.itertuples():
    hb[row.URL] = row.HB
```

```python
driver = None
for url in df['URL']:
    if url in hb:
        continue
        
    if driver is None:
        driver = webdriver.Chrome()
        driver.set_page_load_timeout(30)
        
    try:
        driver.get('https://%s' % url)
        driver.execute_script('return _pbjsGlobals')
        hb[url] = True
        print('%s is hb enabled' % url)
    except TimeoutException as e:
        driver.quit()
        driver = None
        hb[url] = False
        print('%s is not hb enabled' % url)
    except:
        hb[url] = False
        print('%s is not hb enabled' % url)
```

```python
hb_sites = []
for url, is_hb in hb.items():
    hb_sites.append({
        'URL': url,
        'HB': is_hb
    })
    
pd.DataFrame(hb_sites).to_sql('header-bidding-sites', con=get_engine(), index=False, if_exists='append')
```
